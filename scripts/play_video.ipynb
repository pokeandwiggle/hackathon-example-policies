```python
import os
import re

def rename_episodes(base_path):
    # Path containing the episode files
    folder = os.path.join(base_path, "videos/chunk-000/observation.images.rgb_static")

    # Regex to detect episode_00000X.mp4
    pattern = re.compile(r"episode_(\d+)\.mp4")

    # List all episode files
    files = [
        f for f in os.listdir(folder)
        if pattern.match(f)
    ]

    # Sort by numeric episode index
    files.sort(key=lambda x: int(pattern.match(x).group(1)))

    # Rename sequentially
    for new_id, filename in enumerate(files):
        old_path = os.path.join(folder, filename)
        new_filename = f"episode_{new_id:06d}.mp4"
        new_path = os.path.join(folder, new_filename)

        print(f"Renaming {filename} â†’ {new_filename}")
        os.rename(old_path, new_path)

    print("Renaming complete.")


# Example usage:
dataset_name = "my_dataset"
rename_episodes(f"data/{dataset_name}")
```
